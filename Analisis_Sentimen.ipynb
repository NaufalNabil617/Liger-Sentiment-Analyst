# Stopword Removal

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Mengunduh daftar stopwords bahasa Indonesia
stop = stopwords.words('indonesian')

# Menambahkan stopwords tambahan
additional_stopwords = ['nya', 'yg', 'men', 'ya', 'k', 'ok', '']
stop.extend(additional_stopwords)


if 'kurang' in stop:
    stop.remove('kurang')


# Daftar kata yang perlu diperbaiki
string_corrections = {
    'jernihdesainnya': 'jernih desainnya',
    'kuatbassnya': 'kuat bassnya',
    'yasuara': 'ya suara',
    'cemprengbeli': 'cempreng beli',
    'ngambangkurir': 'ngambang kurir',
    'ceptterimakasih': 'cepat terima kasih',
    'bagusbeefungsi': 'bagus berfungsi',
    'baikbarangnya': 'baik barangnya',
    'lamaharganya': 'lama harganya',
    'resmisuaranya': 'resmi suaranya',
    'jernihnyaman': 'jernih nyaman',
    'lamadesainnya': 'lama desainnya',
    'premiumbassnya': 'premium bassnya',
    'mantapharganya': 'mantap harganya',
    'terjangkausuaranya': 'terjangkau suaranya',
    'kemasannyatapi': 'kemasannya tapi',
    'cepatthanks': 'cepat thanks',
    'jernihada': 'jernih ada',
    'resminyaman': 'resmi nyaman',
    'lamakoneksi': 'lama koneksi',
    'kuatdesainnya': 'kuat desainnya',
    'premiumbaterai': 'premium baterai',
    'lamanoise': 'lama noise',
    'terjangkaunggak': 'terjangkau nggak',
    'kresekkreseknyaman': 'kresek-kresek nyaman',
    'mantapsuaranya': 'mantap suaranya',
    'terimakasih': 'terima kasih',
    'diterimapacking': 'diterima packing',
    'lamasuaranya': 'lama suaranya',
    'gurihvalue': 'gurih value',
    'moneyharganya': 'money harganya',
    'terjangkaubassnya': 'terjangkau bassnya',
    'mantapmaterial': 'mantap material',
    'speknysuara': 'spek suara',
    'mantabwarna': 'mantap warna',
    'mantapsesuai': 'mantap sesuai',
    'pesenancepet': 'pesan cepat',
    'nyaterima': 'nya terima',
    'dicobasemoga': 'dicoba semoga',
    'awetthanks': 'awet thanks',
    'packingnyamakasi': 'packingnya makasih',
    'maaaaaannnnttaaaaappppppp': 'mantap',
    'barangnyabagus' : 'barangnya bagus',
    'suarajernihproses' : 'suara jernih proses'
}

regex_corrections = {
    r'\b(?:nggak|ngga|ga|gk|gak)\b': 'tidak',
    r'\b(?:mantap|mantull|mantuu+lu+l)\b': 'mantap',
    r'\b(?:bgt|banget)\b': 'sangat',
    r'\b(?:tidak sesuai)\b': 'buruk',
    r'\b(?:sebelahnya|sebelah|belah)\b': 'cacat',

}

def correct_words_string(text):
    for word, correction in string_corrections.items():
        text = text.replace(word, correction)
    return text

# Fungsi untuk mengganti kata berbasis regex
def correct_words_regex(text):
    for pattern, replacement in regex_corrections.items():
        text = re.sub(pattern, replacement, text, flags=re.IGNORECASE)
    return text

# Fungsi untuk menghapus stopwords
def remove_stopwords(text):
    return ' '.join([word for word in text.split() if word not in stop])

# Gabungkan kedua fungsi dalam pipeline
def preprocess_text(text):
    text = correct_words_string(text)
    text = correct_words_regex(text)
    text = remove_stopwords(text)  # Hapus stopwords
    return text

# Pastikan kolom tidak kosong
data_clean['text_clean'] = data_clean['text_clean'].fillna('')

# Terapkan pipeline pemrosesan
data_clean['text_StopWord'] = data_clean['text_clean'].apply(preprocess_text)

# Tampilkan 50 baris pertama hasil pemrosesan
print(data_clean.head(50))
```
